## Python for Data Science
- Python is a versatile, general-purpose programming language.
- Common libraries covered include NumPy and pandas, which form the core toolset for data scientists.

---
## NumPy Overview
**NumPy** (Numerical Python) is a fundamental python library for numerical computing in Python.

**Core Features:**
- Supports **n-dimensional arrays** (`ndarray`), enabling efficient handling of large data.
- Includes optimized **mathematical functions**: trigonometry, exponents, roots, etc.
- Provides tools for **linear algebra**, **random sampling**, and **vectorized computations**.
- Forms the **foundation** for other libraries like pandas, scikit-learn, TensorFlow.

**Performance Advantages:**
- Stores arrays in **contiguous memory**, unlike Python lists.
- Takes advantage of CPU **caching** and **vector instructions**.
- Mostly implemented in **C**, giving significant speed benefits.

**Applications:**
- Analytical data processing
- Core component in ML model training
- Scientific and numerical simulations

**Basic Functionalities:**
- Creating arrays (from lists, or using generators like `zeros`, `ones`, `random`)
- Reshaping, slicing, and combining arrays
- Statistical functions: mean, median, etc.

## Pandas Overview
**Pandas** is a high-level python library for data manipulation and analysis.

**Core Data Structures:**
- `Series`: One-dimensional labeled array
- `DataFrame`: Two-dimensional structure with labeled rows and columns

**Features:**
- Efficient handling of structured/tabular data
- Methods for missing data handling, grouping, filtering, transforming
- Supports input/output with formats like CSV, Excel, SQL, JSON

**Applications:**
- Cleaning and wrangling unstructured or messy data
- Exploratory data analysis (EDA)
- Time series analysis
- Preprocessing data for machine learning: normalization, feature engineering
